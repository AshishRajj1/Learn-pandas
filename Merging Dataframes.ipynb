```python
import numpy as pd
import pandas as pd
```

```python
students = pd.DataFrame({
    'student_id':[1 , 2 ,3, 4 , 5],
    'name': ['Alice', 'Liam', 'Sophia', 'Mason', 'Emma'],
    'major': ['Computer Science', 'Mathematics', 'Physics', 'Computer Science', 'Chemistry']
})

grades = pd.DataFrame({
    'student_id':[1 , 2 ,3, 6 , 7],
    'credits_earned':[4 , 3 ,2, 4 , 2],
    'gpa': [3.8, 3.5, 3.9, 2.8, 3.7]
})
```

```python
students
```

```python
grades
```

```python
# how='inner': Specifies the type of merge. An inner join returns only the intersecting values (matches) found in both tables    

pd.merge(students,grades, on='student_id', how ='inner')   
```

```python
# how='outer': Specifies that all rows from both tables must be included

pd.merge(students,grades, on='student_id', how ='outer')  
```

```python
# how='left': Specifies that the merge must prioritize the left table.

pd.merge(students,grades, on='student_id', how ='left')
```

```python
# how='left': Specifies that the merge must prioritize the right table.

pd.merge(students,grades, on='student_id', how ='right')
```

Concatination of 2 Dataframes
{stacking multiple DataFrame objects together either vertically (row-wise) or horizontally (column-wise)}


```python
df1 = pd.DataFrame({
'A': ['AO', 'A1', 'A2'],
'B':['B0', 'B1', 'B2'],
'C':['co', 'c1', 'C2']

})

df2 = pd.DataFrame({
'A': ['A3', 'A4', 'A5'],
'B': ['B3', 'B4', 'B5'],
'C': ['C3', 'C4', 'C5']
})
```

```python
df1
```

```python
df2
```

```python
pd.concat([df2 , df1])
```

```python
# axis 1 refers to columns, else it is auto set to axis 0 which refers to rows

pd.concat([df2 , df1], axis = 1)
```

Joining Two DataFrames

```python
df1 = pd.DataFrame({
'name': ['Ace', 'Luffy', 'Sabo']
}, index=[1, 2, 3])

# Second DataFrame
df2 = pd.DataFrame({
'score': [85, 90, 75]
}, index=[2, 3, 4])
```

```python
df1
```

```python
df2
```

```python
df1.join(df2,how='outer')
```

```python
df2.join(df1)
```
