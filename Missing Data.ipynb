```python
import numpy as np
import pandas as pd
```

Finding Missing Data

```python
data = {
    'A': [1, 2, np.nan, 4, 5],
    'B': [1, 2, 3, 4, 5],
    'C': [1, 2, 3, np.nan, np.nan],
    'D': [1, np.nan, np.nan, np.nan, 5]
}
df = pd.DataFrame(data)
df
```

```python
df.isna()
```

```python
# How many Nan(Empty) Values in each Columns
df.isna().sum()
```

```python
# If the Column contains non values or Notification
df.isna().any()
```

Remove Missing Data

```python
df
```

```python
#drops any row containing at least one missing value and returns a new DataFrame copy
df.dropna()
```

```python
df
```

```python
# (thresh=1) stands for "threshold". It sets the minimum number of valid (non-NaN) values a row must have to avoid being dropped.
# By default, it operates on rows (axis=0).
# A row will only be deleted if all of its columns contain NaN or None

df.dropna(thresh=1)
```

Filling the missing Data

```python
df
```

```python
df.fillna(0)
```

```python
# Filling Nun Values with Other Values
values = {'A':0, 'B':100,"C":"Hello", 'D':400}
df.fillna(value=values)
```

```python
# Filling Nun Values with Mean of the Column
df.fillna(df.mean())
```
