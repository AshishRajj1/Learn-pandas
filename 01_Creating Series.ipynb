```python
import numpy as np
import pandas as pd
```

# There are multiple ways to create a pandas Series:

```python
labels = ['x', 'y', 'z']
my_list = [10,20,30]

pd.Series(my_list)
```

```python
pd.Series(my_list,index = labels)
```

```python
arr = np.array([10,20,30])
pd.Series(arr)
```

```python
d = {'a':10, 'b':20,'c':30}
pd.Series(d)
```
